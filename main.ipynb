# 원작이 있는 영화가 더 흥행 할까?

## 1. 주제 선정 이유

&nbsp; 최근 어벤져스 : 엔드게임이 개봉하였다. 어벤져스 엔드게임이 개봉하면서 개봉 초에는 스크린 점유율이 약 80프로였고, 개봉 첫주에만 630만명이 관람 했다. 그리고 현재 개봉 약 3주가 지나가는 지금까지도 예매율 1순위를 지킬 정도로 흥행하고 있다. 흥행의 요인으로는 그동안 쌓아 왔던 마블 영화의 팬층 때문일것이다. 하지만 마블 영화가 아닌 코믹스의 팬층 또한 그 원인이라 생각 한다. 이처럼 원작이 있는 영화들은 대부분 흥행하는 경우가 많다. 해리포터 시리즈,본 시리즈,007 시리즈등 소설을 원작으로해 흥행한 영화는 매우 많았고 웹툰을 원작으로 했었던 우리나라 영화 신과 함께 또한 외국으로 수출할 정도로 흥행 했었다. <br><br>
&nbsp; 반면에 대부분의 영화들은 원작이 없다. 그럼에도 불구하고 흥행을 일으킨다. 예를 들자면 유명한 트랜스포머 시리즈나 아바타, 스타워즈 시리즈는 원작이 없는 영화이다. 이 영화들은 원작이 없음에도 불구하고 엄청난 흥행을 일으켰다. 특히 아바타의 경우 역대 박스 오피스에서 외국 영화중 가장 많은 수익과 관객수를 남긴 영화이다.(5/17 KOBIS 역대 박스오피스 기준) 뿐만 아니라 우리나라의 대부분의 영화들은 원작이 존재하지 않지만 흥행을 한다. <br><br>
&nbsp; 위 사례처럼 원작이 존재하는 영화가 흥행하는 경우도, 그렇지 않은 경우도 있다. 어느쪽이 더 많이 흥행을 할까? 이번 프로젝트를 통해 이를 증명하고자 한다. 

## 2. 가설 정의
### &nbsp; 원작이 있는 영화가 더 흥행 할 것이다. 
&nbsp; 우선, 원작이 있는 영화들은 해당 영화의 순수 팬층이 아닌 원작의 팬층이 존재한다. 그 원작의 팬층들이 영화의 흥행에 큰 몫을 기여 할 것이라고 본다. 또한 해당 영화 개봉 전에 영화의 예고편에서 원작을 두었다고 얘기 하거나 서점이나 온라인 상에서 원작에 대해서도 곧 영화가 개봉한다고 마케팅을 할 것이다. 이를 보고 영화 개봉 전 사람들이 미리 원작을 접하고 흥미를 느껴 영화를 보는 경우도 많을 것이다. 즉, 원작이 없는 경우 보다는 관람객을 더 많이 이끌 수 있는 방법이 많이 존재 한다. 따라서 원작이 있는 영화가 더 흥행 할 것이라 생각한다. 
&nbsp; 이를 증명하기 위해, 우선 원작이 있는 영화 데이터를 가지고 온다. 이를 근 15년 간의 주간 박스오피스 순위를 가지고와, 해당 박스 오피스 순위에 있는 영화중 해당 영화가 있으면 매칭해 박스 오피스 순위 데이터에서 원작이 있는 영화 분류 한 뒤, 그렇지 않은 영화와 총매출액, 총관객수를 비교 한다. 또한 스크린 독점 또한 흥행 요인 중 하나라고 생각한다. 따라서 스크린 횟수와 상영 횟수 대비 관객수값을 비교해 흥행의 요인이 더 많은 스크린 할당, 횟수 할당, 즉 스크린 독점 때문에 관객들의 당시 영화 선택 폭이 줄어서인지를 분석한다.

## 4. 분석을 위한 데이터 가공

#### 1) 원작이 있는 영화 데이터 
- Beautiful Soup 과 Requests 를 사용해 데이터를 획득한다.
- 크롤링 할 url 형식이 일정하지 않았다. 따라서 크롤링 할 url들을 모두 직접 복사해와, dictionary타입의 자료형의 key값으로 저장한다.
- soup.select()시 selector를 사용한다. 이때 각 url 마다 영화 이름들이 다른  selector를 가지고 있었다. 따라서 위에서 선언한 dictonary 타입의 자료형에서 해당 url의 key값에 해당하는 value 값으로 selector를 직접 저장 하였다. <br>
selector example :  ``#mw-pages > div > div > div > ul > li > a`` <br>
- soup.select로 가져온 객체를 한줄씩 가지고 와 string으로 변환한 한다. <br>
 ``<a href="/wiki/%EB%A6%AC%ED%8B%80_%ED%8F%AC%EB%A0%88%EC%8A%A4%ED%8A%B8" title="리틀 포레스트">리틀 포레스트</a>`` <br>
- '>'을 기준으로 split을 해  split 한 결과 값중 1번째 원소를 가지고 와 다시 '<'을 기준으로 split 한다. 마지막 split 결과의 0번째 원소 값이 영화 이름이 된다. 


#### 2) 주간 박스 오피스 데이터 
- 주간/주말 박스 오피스 API를 사용한다. 이때, REST 방식으로 받아 온다. 
- Rest 방식의 응답 형식은 JSON 형태로 받아 온다. <br>
   기본 요청 url :  http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchWeeklyBoxOfficeList.json
- 요청 parameter는 ``{'key': _key, 'targetDt':_tagetDT, 'weekGb':"0"}`` 로 설정 해 GET 방식으로 받아 온다. 이때, weekGB : 0는 주간 데이터를 뜻한다. 
- 요청 parameter의 targetDT는 매일이 아닌 일주일 단위로 들어가야 한다. 따라서 pandas의 daterange를 사용해 2004-01-01 부터 2019-05-12까지 매주 월요일의 날짤를 targetDT로 넘겨 데이터 중복을 막는다.
- salesAcc와 audiAcc는 누적 매출액이므로 매주 갱신 된다. 따라서 갱신된 값을 저장해 준다. scrnCnt와 showCnt의 경우, 해당 일자에 사영된 스크린 횟수, 상영 횟수 이므로 이전의 값에 누적을 시켜준다. 
- 받아온 movieNm, salesAcc,audiAcc,scrnAcc,showCnt는 movie_data에 저장 한다.

#### 3) 데이터의 일관성
- 1번 데이터와 2번 데이터의 출처는 다르다. 따라서 데이터의 일관성이 떨어진다. 두 데이터를 비교하기 위해서는 데이터의 일관성이 필요하다. 따라서 데이터의 일관성을 맞추기 위한 작업을 진행 한다.
    1. 영화 이름에 ``VS.``들어간 경우 : 모두 ``VS``로 바꿔준다. EX) 에이리언 VS 프레데터
    2. 영화 이름에 ``&amp;`` 들어간 경우 : 이는 ``&``로 바꿔준다. EX) 미스터 & 미세스 스미스
    3. 영화 이름에 ``1부, 2부``가 들어간 경우 : 숫자로 대체한다. EX) 해리 포터와 죽음의 성물1
    4. 영화 이름에 (애니)와 같이 추가 설명이 들어간 경우: 추가 설명은 삭제한다.
 
#### 4) 데이터 매칭
- 3번의 과정을 마친 데이터들을 가지고 1번과 2번 데이터 매칭을 시킨다. 
- 데이터 매칭 전에 movie_data에 hasOriginal이라는 원소의 dictionary value값으로 len(movie_data['movieNm']) 만큼 0을 가지고 있어야 한다. 
- 3번 데이터 일관성에서 잡지 못하는 한가지가, 띄어쓰기 문제이다. 예를 들면 '해리 포터'와 '해리포터'는 같은 이름이지만 컴퓨터는 이를 같게 인식 하지 않는다. 따라서 데이터 매칭시 띄어쓰기는 무시하고 매칭한다.
- 띄어쓰기를 무시하고 매칭 하기 위해 영화 이름을 한글자씩 분석 한다. 만일, 모든 문자가 같다면 이는 같은 영화 이름으로 인식 한다.
- 1번의 특정 데이터와 2번의 특정 데이터가 매칭이 되었다면 해당 데이터의 hasOriginal = 1로 만든다.

## 5. 분석 결과 도출

````python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
from matplotlib import rc
import numpy as np

#차트에 한국어를 출력한다
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

#movie_data 파일을 읽어 온다
movielist = pd.read_csv('movie_data.csv', encoding='utf-8')
has_original = movielist[movielist['hasOriginal']==1]
no_original = movielist[movielist['hasOriginal']==0]

print("원작 있는 영화 갯수: ",len(has_original))
print("원작 없는 영화 갯수: ",len(no_original))
````

&nbsp;분석을 위해 ``movie_data.csv`` 에서 읽어온 데이터를 원작 있는 영화와 원작이 없는 영화로 분류 작업을 우선 진행 한다. 분류 작업 결과 원작이 있는 영화가 약 500개, 없는 영화가 약 2000개 정도 나왔다. 해당 영화들 중 상위 20개씩만 가져와 분석을 해도 충분히 가설을 증명 할수 있다고 판단 하였다. 따라서 분석 과정에서는 약 40개의 데이터를 가지고 분석을 진행 한다. 분석의 과정은 다음과 같다.<br>
1. 총 매출액을 기준으로 비교를 한다.
2. 총 관객수를 기준으로 비교를 한다.
3. 총 관객수가 스크린 할당과 상영횟수와 관련이 있는지 조사한다.

&nbsp;``총 매출액``과 ``총 관객수``를 비교 할때는 총 3개의 그래프를 그린다. 첫번째 그래프는 비교 수치 기준 원작이 있는 영화 TOP20 이다. 두번째 그래프는 비교 수치 기준 원작이 없는 영화 TOP20이다. 그리고 마지막 세번째 그래프는 첫번째와 두번째 그래프를 합친것으로, 각 순위에 있는 영화를 비교 하기 위한 그래프이다. 즉, 첫번째와 두번째의 그래프는 TOP20 영화의 이름을 알기 위해서 그린 그래프이고, 세번째 그래프는 각 순위 비교를 위해 그린 그래프 이다. <br><br>
&nbsp;``총 관객수가 스크린 할당과 상영횟수와 관련이 있는지 조사`` 이 그래프는 누적 스크린수 대비 관객수, 누적 상영횟수 대비 관객수의 그래프를 나타내었다. x = 총 누적 스크린수 또는 누적 상영 횟수 그리고 y = 총 관객수로 두어, y = x 를 기준으로 y > x 일때 스크린 할당 또는 상영 횟수가 총 관객수에 큰 영향을 주지 않았다고 판단 할 수 있고, y < x일때, 스크린 독점의 이유등으로 영향을 주었다고 판단 할 수 있다. 

### 1) 총 매출액을 기준으로 비교

````python
#총 매출액 기준으로 정렬한다
has_original = has_original.sort_values(["salesAcc"], ascending=[False])
no_original = no_original.sort_values(["salesAcc"], ascending=[False])

selected_has_original = (has_original[["movieNm","salesAcc"]])[:20].sort_values("salesAcc")
selected_no_original = (no_original[["movieNm","salesAcc"]])[:20].sort_values("salesAcc")


#원작 있는 영화 Top20의 총 매출액을 bar 그래프로 그린다
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)
y = np.arange(20)
rects = plt.barh(y, selected_has_original["salesAcc"], height=0.5)
plt.yticks(y, selected_has_original["movieNm"])
plt.xlim(40000000000,  140000000000)
plt.xlabel('총매출액')
plt.ylabel('영화이름')
plt.title('원작 있는 영화의 top20 총 매출액(0.4 = 40000000000/ 1.4 = 140000000000)')
plt.show()

#원작 없는 영화 Top20의 총 매출액을 bar 그래프로 그린다
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)
y = np.arange(20)
rects = plt.barh(y, selected_no_original["salesAcc"], height=0.5)
plt.yticks(y, selected_no_original["movieNm"])
plt.xlim(40000000000,  140000000000)
plt.xlabel('총매출액')
plt.ylabel('영화이름')
plt.title('원작 없는 영화의 top20 총 매출액(0.4 = 40000000000/ 1.4 = 140000000000)')
plt.show()


#둘의 비교를 위해 영화 이름을 제외 하고 랭크 순으로 비교 한다.
rank = []

for i in range (1,21):
    rank.append(str(21-i)+'위')

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)
y = np.arange(20)
plt.barh(y-0.2, selected_has_original["salesAcc"], label='원작 있는 영화', height=0.4, color='#ff6600')
plt.barh(y+0.2, selected_no_original["salesAcc"], label='원작 없는 영화', height=0.4, color='#0066ff')
plt.yticks(y, rank)
plt.xlim(40000000000,  140000000000)
plt.xlabel('총매출액')
plt.ylabel('영화랭킹')
plt.title('top20 총 매출액 비교 (0.4 = 400억원/ 1.4 = 1400억원)')
plt.show()
````

### 3)  누적 스크린 수 대비 총 관객수, 누적 상영횟수 대비 총 관객수

````python
#총 관객수 기준으로 정렬 후 top40 를 가지고 온다
movielist = movielist.sort_values(["audiAcc"], ascending=[False])
selected_list = (movielist[["movieNm","scrnCnt","audiAcc"]])[:40].sort_values("audiAcc")

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)
value = selected_list["movieNm"].values.tolist()
x = selected_list["scrnCnt"].values.tolist()
y = selected_list["audiAcc"].values.tolist()

#누적 스크린 수 대비 관객수로 그래프를 그린다
for i,type in enumerate(value):
    _x = x[i]
    _y = y[i]
    plt.scatter(_x, _y)
    plt.text(_x+0.3, _y+0.3, type, fontsize=9)
plt.xlabel('누적 스크린 수')
plt.ylabel('관객수')
plt.title('TOP 40 누적 스크린수 대비 관객수')
plt.show()

plt.show()

#총 관객수 기준으로 정렬 후 top40 를 가지고 온다
movielist = movielist.sort_values(["audiAcc"], ascending=[False])
selected_list = (movielist[["movieNm","showCnt","audiAcc"]])[:40].sort_values("audiAcc")

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)
value = selected_list["movieNm"].values.tolist()
x = selected_list["showCnt"].values.tolist()
y = selected_list["audiAcc"].values.tolist()

#누적 상영횟수 대비 관객수로 그래프를 그린다
for i,type in enumerate(value):
    _x = x[i]
    _y = y[i]
    plt.scatter(_x, _y)
    plt.text(_x+0.3, _y+0.3, type, fontsize=9)
plt.xlabel('누적 상영횟수')
plt.ylabel('관객수')
plt.title('TOP 40 누적 상영횟수 대비 관객수')
plt.show()

plt.show()
````

## 6. 결론
&nbsp; 처음에 설정한 가설 ``원작이 있는 영화가 더 흥행 할 것이다``을 증명 하기 위해 원작이 있는 영화 데이터와, 근 15년간의 주간 박스 오피스 데이터를 가지고와 역대 박스 오피스 데이터로 재구성해 원작 있는 영화 그리고 없는 영화 TOP 20를 분석 하였다. 그 결과 해당 가설은 옳지 않음을 알 수 있었다. 위의 분석 결과를 정리 하면 다음과 같다.<br>
1. 원작이 없는 영화의 경우 총 매출액이 약 200억정도 더 많음
2. 원작이 없는 영화의 총 관객수 또한 약 500만 정도 더 많음
3. 스크린 독점 또한 원작이 있는 영화 보다 심하지 않아 스크린 독점이 관객수에 크게 영향을 미치지 않았음

&nbsp; 해당 가설이 옳지 않았다는 결론이 나오게 된 이유를 생각해 보면 다음과 같다고 생각 할 수 있다. 우선 원작이 없는 영화 TOP20의 경우 대부분 한국 영화이다. 한국 영화의 경우 시사회, 방송, 배우들의 무대 인사등 다양한 방법으로 마케팅을 많이 한다. 그리고 입소문을 많이 타는 경향이 있다. 이 때문에 외국 영화보다 많은 관객수를 끌어들일 수 있었던 것 같다고 생각 한다. <br>
&nbsp; 또한 해당 데이터를 얻어오는 시점의 문제가 있었다고 생각한다. 해당 데이터를 얻어온 시점은 ``pd.date_range("2004-01-01","2019-05-12",freq='W-MON')`` 이다. 즉, 원작이 있는 영화의 1위 영화인 어벤져스 : 엔드게임이 상영하고 있었던 기간이다. 원작 있는 영화의 1위 데이터가 원작 없는 1위 데이터와 다르게 완전한 데이터가 아니라는 것이다. 이를 감안 하고 분석한 결과이다.
